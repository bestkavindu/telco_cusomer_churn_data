```python
yes_no_mapping = {'Yes':1, 'No':0}
data['Partner'] =data['Partner'].map(yes_no_mapping)
data['Dependents'] =data['Dependents'].map(yes_no_mapping)
data['PhoneService'] =data['PhoneService'].map(yes_no_mapping)
data
```

```python
InternetServiceMapping = {'Yes':1, 'No':0, 'No internet service':0}
data['OnlineSecurity'] = data['OnlineSecurity'].map(InternetServiceMapping)
data['OnlineBackup'] = data['OnlineBackup'].map(InternetServiceMapping)
data['DeviceProtection'] = data['DeviceProtection'].map(InternetServiceMapping)
data['TechSupport'] = data['TechSupport'].map(InternetServiceMapping)
data['StreamingTV'] = data['StreamingTV'].map(InternetServiceMapping)
data['StreamingMovies'] = data['StreamingMovies'].map(InternetServiceMapping)
data
```

```python
PhoneServiceMapping = {'Yes':1, 'No':0, 'No phone service':0}
data['MultipleLines'] = data['MultipleLines'].map(PhoneServiceMapping)
data
```

```python
DSLFiberMapping = {'DSL':1, 'Fiber optic':1, 'No':0}
data['InternetService'] = data['InternetService'].map(DSLFiberMapping)
data
```

```python
ContractMapping = {'Month-to-month':'MM', 'One year':'OY', 'Two year':'TY'}
data['Contract'] = data['Contract'].map(ContractMapping)
data
```

```python
data['PaperlessBilling'] = data['PaperlessBilling'].map(yes_no_mapping)
data
```

```python
PaymentMethodMapping = {'Electronic check':'check', 'Mailed check':'check', 'Bank transfer (automatic)':'bank', 'Credit card (automatic)':'card'}
data['PaymentMethod'] = data['PaymentMethod'].map(PaymentMethodMapping)
data
```

```python
data['Churn'] = data['Churn'].map(yes_no_mapping)
data
```

```python
sex =pd.get_dummies(data['gender'],drop_first=True)
sex
```

```python
contract = pd.get_dummies(data['Contract'], drop_first=True)
paymethod = pd.get_dummies(data['PaymentMethod'], drop_first=True)
```

```python
contract
```

```python
paymethod
```

```python
data = pd.concat([data,sex,contract,paymethod], axis=1)
data
```

```python
data.drop(['customerID','gender','Contract','PaymentMethod'], axis=1, inplace=True)
data
```

```python
from sklearn.preprocessing import LabelEncoder
lblEnc = LabelEncoder()
data_tf = data.copy()

for i in data.columns:
    data_tf[i] = lblEnc.fit_transform(data[i])
```

```python
data_tf
```

```python
X = data_tf.drop('Churn', axis=1)
Y = data_tf['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, '------> Traning feature shape')
print(X_test.shape, '------> Testing feature shape')
print(Y_train.shape, '------> Traning labels shape')
print(Y_test.shape, '------> Testings labels shape')
```

```python
 from sklearn.linear_model import LogisticRegression
```

```python
data_log_model = LogisticRegression()
```

```python
data_log_model.fit(X_train, Y_train)
```

```python
data_pred = data_log_model.predict(X_test)
```

```python
data_pred
```

```python
 from sklearn.metrics import classification_report
 print(classification_report(Y_test, data_pred))
```

```python
import sklearn.metrics as sm

```

```python
accurace_of_data = sm.accuracy_score(Y_test, data_pred)
```

```python
print('Acuracy scor == ', str(accurace_of_data))
```
